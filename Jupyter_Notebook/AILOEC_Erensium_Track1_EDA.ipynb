```python
#Importing Library's

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Hackathon_IITII/Juputer_Files/Track_1.csv")
```

```python
# Droping the Column which are not useful

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df = df.drop(['1=defaulted'],axis=1)
```

```python
df.head()
```

```python
#Let's First See Data In our prediction class

sns.countplot(x ="Loan Defaulted or not" , data = df, palette = ['lightgreen','lightblue'])
```

### $\color{red}{\text{*Note:- }}$

We Can clear see here the data is imbalanced the minority class is the class 2. Which contains almost half the sample as compared to main class so in future we will be applying some balancing technique like undersampling,oversampling,SMOT to balance the data
